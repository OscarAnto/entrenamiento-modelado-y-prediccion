# Se separa en 80/20 el dataset para el entrenamiento y para hacer la prueba con el algoritmo

```python
ml_dataset_copy = dataframe.copy()
train = ml_dataset_copy.sample(frac = 0.80, random_state = 0)
test = ml_dataset_copy.drop(train.index)

print('Train data has %i rows and %i columns' %(train.shape[0], train.shape[1]))
print('Test data has %i rows and %i columns' % (test.shape[0], test.shape[1]))
```

```python
drop_rows_when_missing = []
impute_when_missing = [{'impute_with': u'MEAN', 'feature': u'R707_Quantity_sum'}, {'impute_with': u'MEAN', 'feature': u'R364_Quantity_sum'}, {'impute_with': u'MEAN', 'feature': u'R565_Quantity_sum'}, {'impute_with': u'MEAN', 'feature': u'R044_Quantity_sum'}, {'impute_with': u'MEAN', 'feature': u'nb_km'}, {'impute_with': u'MEAN', 'feature': u'R783_Quantity_sum'}, {'impute_with': u'MEAN', 'feature': u'R193_Quantity_sum'}, {'impute_with': u'MEAN', 'feature': u'R446_Quantity_sum'}, {'impute_with': u'MEAN', 'feature': u'Quantity_sum'}, {'impute_with': u'MEAN', 'feature': u'Time_begin_exploitation'}, {'impute_with': u'MEAN', 'feature': u'Initial_km'}, {'impute_with': u'MEAN', 'feature': u'R575_Quantity_sum'}, {'impute_with': u'MEAN', 'feature': u'R064_Quantity_sum'}, {'impute_with': u'MEAN', 'feature': u'R119_Quantity_sum'}, {'impute_with': u'MEAN', 'feature': u'R396_Quantity_sum'}, {'impute_with': u'MEAN', 'feature': u'R606_Quantity_sum'}, {'impute_with': u'MEAN', 'feature': u'R417_Quantity_sum'}]
# Features for which we drop rows with missing values"
#borrar casillas vacias
for feature in drop_rows_when_missing:
    train = train[train[feature].notnull()]
    test = test[test[feature].notnull()]
    print('Dropped missing records in %s' % feature)
#llena campos vacios input when missing   
# Features for which we impute missing values"
for feature in impute_when_missing:
    if feature['impute_with'] == 'MEAN':
        v = train[feature['feature']].mean()
    elif feature['impute_with'] == 'MEDIAN':
        v = train[feature['feature']].median()
    elif feature['impute_with'] == 'CREATE_CATEGORY':
        v = 'NULL_CATEGORY'
    elif feature['impute_with'] == 'MODE':
        v = train[feature['feature']].value_counts().index[0]
    elif feature['impute_with'] == 'CONSTANT':
        v = feature['value']
    train[feature['feature']] = train[feature['feature']].fillna(v)
    test[feature['feature']] = test[feature['feature']].fillna(v)
    print('Imputed missing values in feature %s with value %s' %(feature['feature'], coerce_to_unicode(v)))
```

```python
rescale_features = {u'R707_Quantity_sum': u'AVGSTD', u'Quantity_sum': u'AVGSTD', u'R364_Quantity_sum': u'AVGSTD', u'R565_Quantity_sum': u'AVGSTD', u'R044_Quantity_sum': u'AVGSTD', u'R783_Quantity_sum': u'AVGSTD', u'R193_Quantity_sum': u'AVGSTD', u'R446_Quantity_sum': u'AVGSTD',u'nb_km': u'AVGSTD', u'Time_begin_exploitation': u'AVGSTD', u'R606_Quantity_sum': u'AVGSTD', u'Initial_km': u'AVGSTD', u'R575_Quantity_sum': u'AVGSTD', u'R064_Quantity_sum': u'AVGSTD', u'R119_Quantity_sum': u'AVGSTD', u'R396_Quantity_sum': u'AVGSTD', u'R417_Quantity_sum': u'AVGSTD'}
for (feature_name, rescale_method) in rescale_features.items():
    if rescale_method == 'MINMAX':
        _min = train[feature_name].min()
        _max = train[feature_name].max()
        scale = _max - _min
        shift = _min
    else:
        shift = train[feature_name].mean()
        scale = train[feature_name].std()
    if scale == 0.:
        del train[feature_name]
        del test[feature_name]
        print('Feature %s was dropped because it has no variance' % feature_name)
    else:
        print('Rescaled %s' % feature_name)
        train[feature_name] = (train[feature_name] -shift).astype(np.float64) / scale
        test[feature_name] = (test[feature_name] - shift).astype(np.float64) / scale
```
